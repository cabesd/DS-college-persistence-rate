```python
comb['Area'] = comb['Area'].astype(object)
```

## Drop features with > 0.85  empty

```python
nans = summary_missing_data(comb)
drop = list(nans[ nans['Percent'] > 0.85 ].index)
comb.drop(columns=drop, inplace=True)
comb.shape
```

## Fill the others

```python
nans = summary_missing_data(comb)
nans
```

```python
comb[list(nans.index)].dtypes
```

```python
cat_cols = list(nans[comb[list(nans.index)].dtypes == 'object'].index)
num_cols = list(nans[comb[list(nans.index)].dtypes == 'float64'].index)
```

```python
# vals_by_area = comb.groupby('Area')[list(nans.index)].mean()
# print(vals_by_area.shape[0], ' Areas and ', vals_by_area.shape[1], 'Missing featues')
# vals_by_area.head()
```

```python
cat_cols
```

```python
for c in num_cols:
    print(c)
    comb[c] = comb.groupby('Area').transform(lambda x: x.fillna(x.mean()))
```

```python
summary_missing_data(comb)
```

```python
for c in cat_cols:
    print(c)
    comb[c] = comb.groupby('Area').transform(lambda x: x.fillna(x.mode()))
```
